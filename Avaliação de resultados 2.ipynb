```python
def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)

plt.figure(figsize =(12, 7))

bpl = plt.boxplot(AcuraciaTe, positions=np.array(range(len(AcuraciaTe.T)))*2.0-0.4, sym='', widths=0.7)
bpr = plt.boxplot(AcuraciaTr, positions=np.array(range(len(AcuraciaTr.T)))*2.0+0.4, sym='', widths=0.7)
set_box_color(bpl, '#D7191C') # colors are from http://colorbrewer2.org/
set_box_color(bpr, '#2C7BB6')

# draw temporary red and blue lines and use them to create a legend
plt.plot([], c='#D7191C', label='Teste')
plt.plot([], c='#2C7BB6', label='Treino')
plt.legend()

plt.xticks(range(0, len(metleg) * 2, 2), metleg)
plt.xlim(-2, len(metleg)*2)
plt.title("Acurácia")
plt.xlabel("Método")
plt.ylabel("Percentual")

plt.savefig('Imagens2\Avaliacao\BPAcuracia.jpeg')
```
